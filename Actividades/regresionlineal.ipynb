```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
np.random.seed(0)
horas_estudio = np.random.rand(100, 1) * 10
calificaciones = 2 * horas_estudio + np.random.randn(100, 1) * 2
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(horas_estudio, calificaciones, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(train_data, train_labels)
```

```python
predictions = model.predict(test_data)
mse = mean_squared_error(test_labels, predictions)
print("Error Cuadrático Medio (MSE):", mse)
```

```python
plt.scatter(test_data, test_labels, color='blue', label='Datos de Prueba')
plt.plot(test_data, predictions, color='green', linewidth=3, label='Modelo de Regresión Lineal')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación')
plt.title('Regresión Lineal: Horas de Estudio vs. Calificación')
plt.legend()
plt.show()
```
